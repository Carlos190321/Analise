```python
df = pd.read_csv("C:/Users/User/Downloads/trabalho.csv", delimiter=';')
df.rename(columns={'customerID': 'ID_Ùnico', 'PhoneService': 'Serviço_Telefônico',                 
                   'StreamingTV': 'TransmissãoTV', 'PaymentMethod': 'Forma_de_Pagamento', 'Churn': 'Abandono'}, inplace=True)


Tradução = {
    'Yes': 'Sim',
    'No': 'Não',
    'Female': 'Fêmea',
    'Male': 'Macho',
    'Fiber optic': 'Fibra_óptica',
    'Month-to-month': 'Mês_a_mês',
    'One year': 'Um_ano'}
df = df.replace(Tradução)


tradução_forma_pagamento = {'Electronic check': 'Cheque Eletrônico',
    'Mailed check': 'Cheque Enviado',
    'Bank transfer (automatic)': 'Transferência Bancária Automática',
    'Credit card (automatic)': 'Cartão de Crédito Automático'}
df['Forma_de_Pagamento'] = df['Forma_de_Pagamento'].map(tradução_forma_pagamento)

tradução_Tipo_Contrato = {'Month-to-month': 'Mês_a_mês'}
pd.set_option('future.no_silent_downcasting', True)

df = df.replace(tradução_Tipo_Contrato).infer_objects(copy=False)
df.head(10)
```
